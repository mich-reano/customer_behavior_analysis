```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('customer_shopping_behavior.csv')
df.head()
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
df.info()
```

```python
# create a new column age_group
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
df[['age', 'age_group']].head(10)
```

```python
df['frequency_of_purchases'].unique()
```

```python
# create column purchase_frequency_days
frequency_mapping = {'Fortnightly': 14, 'Weekly': 7, 'Monthly': 30, 'Quarterly': 90, 'Bi-Weekly': 14, 'Annually': 365, 'Every 3 Months': 90}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
df[['frequency_of_purchases', 'purchase_frequency_days']].head(10)
```

```python
df[['promo_code_used', 'discount_applied']].tail(10)
```

```python
df = df.drop(columns=['promo_code_used'], axis=1)
df.columns
```

```python
import pymysql

conn = pymysql.connect(
    host="localhost",
    user="root",
    password="@michreano",
    database="customer_behavior"
)

cursor = conn.cursor()

# Auto-generate CREATE TABLE from DataFrame columns
columns = df.columns
dtypes = df.dtypes

sql_cols = []

for col, dtype in zip(columns, dtypes):
    if "int" in str(dtype):
        sql_type = "INT"
    elif "float" in str(dtype):
        sql_type = "FLOAT"
    else:
        sql_type = "VARCHAR(255)"   # default for strings
        
    sql_cols.append(f"`{col}` {sql_type}")

create_table_sql = f"""
CREATE TABLE IF NOT EXISTS customer (
    {", ".join(sql_cols)}
)
"""

cols = ",".join(df.columns)
placeholders = ",".join(["%s"] * len(df.columns))

sql = f"INSERT INTO customer ({cols}) VALUES ({placeholders})"

cursor.execute(create_table_sql)
cursor.executemany(sql, df.values.tolist())
conn.commit()
conn.close()

print(f"Data succesfully inserted into the database.")
```
